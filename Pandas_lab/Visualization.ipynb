```python
### Multivariate distribution (scatter plots)
mean, cov = [-1, 1], [(0.8, .5), (0.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])


plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
plt.scatter(df["x"], df["y"])

plt.subplot(1,3,2)
plt.hist(df["x"])
plt.title("X histogram")

plt.subplot(1,3,3)
plt.title('Y histogram')
plt.hist(df["y"])
```

### Bokeh

```python
# Example

# import high level plotting function
from bokeh.plotting import figure, output_file, show

# Create/Get data to plot
x = [1, 2, 3, 4, 5]
y = [10, 5, 12, 8, 20]

# Location of the javascript graph
output_file("bokeh.html")

# Build the graph by operating on the figure object
p = figure(title="small example")
p.line(x, y)

# This will open a new tab showing bokeh.html file
show(p)
```

```python
from bokeh.io import output_notebook
output_notebook()

x = [1, 2, 3, 4, 5]
y = [10, 5, 12, 8, 20]

p = figure(title="small example")
p.line(x, y)
show(p)
```

#### Line Plot

```python
x = np.arange(0,5,0.1)
xsqr = x ** 2
xcube = x ** 3

output_notebook()

# create a new plot
p = figure(
   tools="pan,box_zoom,reset,save", # Toolbox to the right of the plot
   title="Bokeh example",
   x_axis_label='X', y_axis_label='f(X)'
)

p.line(x, x, legend="f(x)=x")
p.line(x, xsqr, legend="f(x) = x^2", line_width=3) # Note line_width
p.line(x, xcube, legend="f(x)=x^3", line_color="red") #Note line color

show(p)
```

#### Markers

```python
x = np.arange(0,5,0.2)
# create a new plot
p = figure(plot_height=500, plot_width=500)
p.circle(x, x**2, size=x)
p.square(x, x**3, size=x, color='red')
p.line(x, x**3)
show(p)
```

#### Stacked Bars

```python
df = pd.DataFrame({
    '2016': {'Digital': 800, 'Print': 3000},
    '2017': {'Digital': 2200, 'Print': 2800},
    '2018': {'Digital': 3400, 'Print': 2000}
}).transpose()
df
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

p = figure(
    title='Advertising spend by year',
    x_range=df.index.values # note the use of .values
)
p.vbar(x=df.index, width=0.8, bottom=0, top=df['Digital'], color='red')
p.vbar(x=df.index, width=0.8, bottom=df['Print'], top=df['Digital'], color='green')
show(p)
```

#### Pie Charts

```python
from bokeh.plotting import figure, show
from numpy import pi

# Mark the start and end of each section
percents = [0, 0.3, 0.4, 0.6, 0.9, 1]
starts = [p*2*pi for p in percents[:-1]]
ends = [p*2*pi for p in percents[1:]]
colors = ["red", "green", "blue", "orange", "yellow"]

p = figure()
p.wedge(x=0, y=0, radius=1, start_angle=starts, end_angle=ends, color=colors)
show(p)
```

#### Sub Plots

```python
from bokeh.io import output_file, show, output_notebook
from bokeh.layouts import column, row
from bokeh.plotting import figure

output_notebook()
x  = np.arange(0,5,0.1)
# Note the explicit use of plot_width and plot_height
p1 = figure(title='square', plot_width = 300, plot_height = 300)
p2 = figure(title='cube', plot_width=300, plot_height = 300)

p1.line(x, x ** 2, color='red')
p2.line(x, x ** 3, color='blue')
```

```python
show(row(p1, p2))
```
